```python
import math

# -------------------------------
# SAMPLE DATA (replace with dataset if needed)
# -------------------------------
# Features: [Study hours, Attendance, Parent Edu, Test Prep, Sleep]
X = [
    [2, 75, 2, 0, 6],
    [4, 85, 3, 1, 7],
    [3, 80, 2, 1, 6],
    [5, 90, 4, 1, 8],
    [1, 70, 1, 0, 5]
]

# Target: Final Score
Y = [60, 75, 70, 85, 55]

# -------------------------------
# ADD BIAS TERM (X0 = 1)
# -------------------------------
for i in range(len(X)):
    X[i] = [1] + X[i]   # adding intercept term

# -------------------------------
# MATRIX FUNCTIONS
# -------------------------------
def transpose(matrix):
    return list(map(list, zip(*matrix)))

def multiply(A, B):
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

def inverse(matrix):
    # Only for 2x2 (for simplicity in demo)
    det = matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]
    return [
        [matrix[1][1]/det, -matrix[0][1]/det],
        [-matrix[1][0]/det, matrix[0][0]/det]
    ]

# -------------------------------
# CONVERT Y TO COLUMN MATRIX
# -------------------------------
Y_mat = [[y] for y in Y]

# -------------------------------
# NORMAL EQUATION: B = (XᵀX)^(-1) XᵀY
# -------------------------------
X_T = transpose(X)
XTX = multiply(X_T, X)

# NOTE: Full inverse is complex; skipping for large matrices
# Instead using simple gradient descent below
# -------------------------------

# -------------------------------
# GRADIENT DESCENT IMPLEMENTATION
# -------------------------------
def predict(X, beta):
    return sum([X[i]*beta[i] for i in range(len(beta))])

def gradient_descent(X, Y, lr=0.0001, epochs=10000):
    n = len(X)
    m = len(X[0])  # number of features
    beta = [0] * m

    for _ in range(epochs):
        gradients = [0] * m

        for i in range(n):
            y_pred = predict(X[i], beta)
            error = y_pred - Y[i]

            for j in range(m):
                gradients[j] += error * X[i][j]

        for j in range(m):
            beta[j] -= (lr * gradients[j]) / n

    return beta

# -------------------------------
# TRAIN MODEL
# -------------------------------
beta = gradient_descent(X, Y)

print("Model Coefficients:")
print(beta)

# -------------------------------
# TEST PREDICTION
# -------------------------------
test = [1, 4, 85, 3, 1, 7]  # new student data
prediction = predict(test, beta)

print("Predicted Score:", round(prediction, 2))
```

```python
import csv

# -------------------------------
# FILE PATH
# -------------------------------
file_path = "/content/auto-mpg.csv"

X = []
Y = []

# -------------------------------
# LOAD DATA
# -------------------------------
with open(file_path, 'r') as file:
    reader = csv.DictReader(file)

    for row in reader:
        try:
            mpg = float(row['mpg'])
            horsepower = row['horsepower']

            # Handle missing values ("?")
            if horsepower == '?':
                continue

            horsepower = float(horsepower)

            # Polynomial features: x, x^2, x^3
            features = [1, horsepower, horsepower**2, horsepower**3]

            X.append(features)
            Y.append(mpg)

        except:
            continue

# -------------------------------
# PREDICTION FUNCTION
# -------------------------------
def predict(x, beta):
    return sum(x[i] * beta[i] for i in range(len(beta)))

# -------------------------------
# GRADIENT DESCENT
# -------------------------------
def gradient_descent(X, Y, lr=0.0000000001, epochs=10000): # Reduced learning rate further
    n = len(X)
    m = len(X[0])
    beta = [0] * m

    for _ in range(epochs):
        gradients = [0] * m

        for i in range(n):
            y_pred = predict(X[i], beta)
            error = y_pred - Y[i]

            for j in range(m):
                gradients[j] += error * X[i][j]

        for j in range(m):
            beta[j] -= (lr * gradients[j]) / n

    return beta

# -------------------------------
# TRAIN MODEL
# -------------------------------
beta = gradient_descent(X, Y)

print("Model Coefficients:")
print(beta)

# -------------------------------
# TEST PREDICTION
# -------------------------------
# Example: horsepower = 100
hp = 100
test = [1, hp, hp**2, hp**3]

prediction = predict(test, beta)

print("Predicted MPG for horsepower =", hp, ":", round(prediction, 2))
```
